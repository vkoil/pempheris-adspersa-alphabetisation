```python
import pandas as pd
import numpy as np

import librosa
import librosa.display
import librosa.segment

import IPython.display

import matplotlib.pyplot as plt
```

```python
# loads an audio file as an array of amplitudes (integers)
sample, sr = librosa.load("data/samples/001_021530_021630.wav",sr=44000)
```

```python
click_profile = sample[int(sr*7.40):int(sr*7.80)]
noise_profile = sample[int(sr*0.15):int(sr*0.55)]
```

```python
IPython.display.Audio(data=click_profile, rate=sr)
```

```python
IPython.display.Audio(data=noise_profile, rate=sr)
```

```python
# displays waveform
librosa.display.waveshow(click_profile, sr=sr)
```

```python
librosa.display.waveshow(noise_profile, sr=sr)
```

```python
stftarrayclick=librosa.stft(click_profile, n_fft=256, hop_length=8, win_length=256, window='hann', center=True, dtype=None, pad_mode='constant')

fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(stftarrayclick),y_axis='log', x_axis = 'time')
ax.set_title('Power spectrogram click')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```

```python
stftarraynoise=librosa.stft(noise_profile, n_fft=256, hop_length=10, win_length=256, window='hann', center=True, dtype=None, pad_mode='constant')

fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(stftarraynoise),y_axis='log', x_axis = 'time')
ax.set_title('Power spectrogram noise')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```
