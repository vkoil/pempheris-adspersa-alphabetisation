```python
import pandas as pd
import numpy as np

import librosa
import librosa.display
import librosa.segment
import IPython.display
import matplotlib.pyplot as plt
```

```python
# loads an audio file as an array of amplitudes (integers)
sample, sr = librosa.load("C:/Users/Isabe/OneDrive/Documents/GitHub/bigeye_vocalisation_recognition/data/samples/001_021530_021630.wav",sr=44000)
```

```python
# displays waveform
librosa.display.waveshow(sample[int(sr*7.58):int(sr*7.59)], sr=sr)
```

```python
librosa.display.waveshow(sample[int(sr*0.15):int(sr*0.16)], sr=sr)
```

```python
click = np.abs(librosa.stft(sample[int(sr*7.58):int(sr*7.68)]))
```

```python
noise = np.abs(librosa.stft(sample[int(sr*0.15):int(sr*0.25)]))
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(click),y_axis='log',x_axis='time')
ax.set_title('Power spectrogram')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(noise),y_axis='log',x_axis='time')
ax.set_title('Power spectrogram')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(stftarrayclick),y_axis='log', x_axis = 'time')
ax.set_title('Power spectrogram click')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(stftarraynoise),y_axis='log', x_axis = 'time')
ax.set_title('Power spectrogram noise')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```
